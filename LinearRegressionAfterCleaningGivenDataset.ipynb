```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading Data
data = pd.read_csv('share.csv')
data.head()
```

```python
#Replacing the * with NaN value in the given dataset
data.replace('*', np.NaN, inplace = True)
data.replace('**', np.NaN, inplace = True)
data.head()
```

```python
print(len(data))
```

```python
#Dropping the rows having non - NaN values in less than 3 columns
data.dropna(inplace = True, thresh = 3)
```

```python
print(len(data))
```

```python
data.head(len(data))
```

```python
#Replacing the NaN values with the mean of the respective column
data['Recent 5-Year Trend (2) in Death Rates'] = data['Recent 5-Year Trend (2) in Death Rates'].apply(lambda x : float(x))
data['Recent 5-Year Trend (2) in Death Rates'].replace(np.NaN, data['Recent 5-Year Trend (2) in Death Rates'].mean(), inplace = True)
data['Recent 5-Year Trend (2) in Death Rates'].head(len(data))
```

```python
#Replacing the NaN values with the mean of the respective column
data['Lower 95% Confidence Interval for Trend'] = data['Lower 95% Confidence Interval for Trend'].apply(lambda x : float(x))
data['Lower 95% Confidence Interval for Trend'].replace(np.NaN, data['Lower 95% Confidence Interval for Trend'].mean(), inplace = True)
data['Lower 95% Confidence Interval for Trend'].head(len(data))
```

```python
#Replacing the NaN values with the mean of the respective column
data['Upper 95% Confidence Interval for Trend'] = data['Upper 95% Confidence Interval for Trend'].apply(lambda x : float(x))
data['Upper 95% Confidence Interval for Trend'].replace(np.NaN, data['Upper 95% Confidence Interval for Trend'].mean(), inplace = True)
data['Upper 95% Confidence Interval for Trend'].head(len(data))
```

```python
data.head(len(data))
```

```python
X = data['Recent 5-Year Trend (2) in Death Rates'].values
Y = data['Lower 95% Confidence Interval for Trend'].values
```

```python
#Mean x and y
mean_x = np.mean(X)
mean_y = np.mean(Y)
print("Mean x = ", mean_x)
print("Mean y = ", mean_y)
```

```python
#Calculating b0 and b1
num = 0
den = 0
for i in range(len(X)):
    num += (X[i] - mean_x) * (Y[i] - mean_y)
    den += (X[i] - mean_x) ** 2
b1 = num / den
b0 = mean_y - (b1 * mean_x)
print("b1 = ", b1, "b0 = ", b0)
```

```python
#Plotting Values
max_x = np.max(X) + 20
min_x = np.min(X) - 20

x = np.linspace(min_x, max_x, 1000)
y = b0 + b1 * x

plt.plot(x, y, color = "blue", label = "Regression Line")
plt.scatter(X, Y, color = "red", label = "Scatter Point")
plt.xlabel("Head size in cm^3")
plt.ylabel("Brain Weight(grams)")
plt.legend()
plt.show()
```

```python
#Calculating Error
rmse = 0
for i in range(len(X)):
    y_pred = b0 + b1 * X[i]
    rmse += (Y[i] - y_pred) ** 2
rmse = np.sqrt(rmse / len(X))
print("RMSE = ", rmse)
```
